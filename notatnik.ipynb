```python
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
#from PyQt5.QtWidgets import QLabel
# Filename: g_layout.py


class Kalkulator(QMainWindow):
    files = ""
    def __init__(self):
        super(Kalkulator, self).__init__()
        self.initUI()
        
    def openFile(self):
        files, _ = QFileDialog.getOpenFileName(self,"QFileDialog.getOpenFileNames()")
        s = open(files,"rt")
        with s:
            r = s.read()
            self.textEdit2.setText(r)
        self.status.showMessage("Otwarto plik")
        
    def saveFile(self):
        files, _ = QFileDialog.getSaveFileName(self, 'Save File')
        file = open(files,'w')
        text = self.textEdit2.toPlainText()
        file.write(text)
        file.close()
        self.status.showMessage("Zapisano plik")
            
    def newFile(self):
        self.status.showMessage("Stworzono plik")
        
    def undo(self):
        self.textEdit2.undo()
        self.status.showMessage("Cofnieto")
        
    def redo(self):
        self.textEdit2.redo()
        self.status.showMessage("Powtorzono")
        
    def cut(self):
        self.textEdit2.cut()
        self.status.showMessage("Wycieto")
        
    def copy(self):
        self.textEdit2.copy()
        self.status.showMessage("Skopiowano")
        
    def paste(self):
        self.textEdit2.paste()
        self.status.showMessage("Wklejono")
        
    def changeSize(self):
        index = self.combo.currentIndex()
        size = self.combo.itemData(index)
        self.textEdit2.setFontPointSize(size)
    
    def changeColor(self, i, j):
        self.textEdit2.setStyleSheet("background-color: "+colors[i][j])
       

    def initUI(self):
        Widget = QWidget()
        self.setCentralWidget(Widget)
       
        grid = QGridLayout()
        
        #grid na duze okna
        oknawidget = QWidget()
        oknagrid = QGridLayout()
        self.textEdit2 = QTextEdit()
        #combo
        self.combo = QComboBox()
        self.combo.addItem("12", 12)
        self.combo.addItem("14", 14)
        self.combo.addItem("16", 15)
        self.combo.setCurrentText("12")
        oknagrid.addWidget(self.combo,0,0,1,1)
        #radiobuttony
        widget_na_radio = QWidget()
        grid_na_radio = QGridLayout()
        self.radio1 = QRadioButton("Times New Roman")
        self.radio1.setChecked(True)
        self.radio2 = QRadioButton("Arial")
        self.radio3 = QRadioButton("Courier New")
        grid_na_radio.addWidget(self.radio1,0,0)
        grid_na_radio.addWidget(self.radio2,1,0)
        grid_na_radio.addWidget(self.radio3,2,0)
        widget_na_radio.setLayout(grid_na_radio)
        oknagrid.addWidget(widget_na_radio,1,0,1,1)
        
        #kolorowe przyciski
        widget_na_przyciski = QWidget()
        grid_na_przyciski = QGridLayout()
        colors = [["Maroon", "Red", "Orange", "Yellow"], ["Olive", "Green", "Purple", "Fuchsia"]\
                 ,["Lime", "Teal", "Aqua", "Blue"], ["Navy", "Gray", "Silver", "Black"]]
        for i in range(0,4):
            for j in range(0,4):
                b = QPushButton()
                b.setStyleSheet("background-color: " + colors[i][j])
                grid_na_przyciski.addWidget(b,i,j)
        widget_na_przyciski.setLayout(grid_na_przyciski)
       
        oknagrid.addWidget(widget_na_przyciski,2,0,1,1)
        oknagrid.addWidget(self.textEdit2,0,2,4,1)
        oknawidget.setLayout(oknagrid)
        grid.addWidget(oknawidget,1,0)
        
        #Menu
        exitAction = QAction('Exit', self)
        exitAction.setShortcut('Ctrl+Q')
        exitAction.setStatusTip('Exit application')
        exitAction.triggered.connect(self.close)
        #otworz
        otworz = QAction('Open', self)
        otworz.setStatusTip('Open file')
        otworz.triggered.connect(self.openFile)
        #zapisz
        otworz = QAction('Open', self)
        otworz.setStatusTip('Open file')
        otworz.triggered.connect(self.openFile)
       
        self.combo.currentIndexChanged.connect(self.changeSize)
        self.radio1.toggled.connect(lambda:self.textEdit2.setFontFamily("Arial"))
        self.radio2.toggled.connect(lambda:self.textEdit2.setFontFamily("Calibri"))
        self.radio3.toggled.connect(lambda:self.textEdit2.setFontFamily("Georgia"))
        
        
        #self.textEdit = QTextEdit()
        #self.textEdit.setFontFamily("Arial")
        #self.textEdit.setFontPointSize(12)
        self.status = self.statusBar()
        exitAction = QAction('Koniec', self)
        exitAction.setShortcut('Ctrl+Q')
        exitAction.triggered.connect(self.close)        
        
        newFileAction = QAction("Nowy plik", self)
        newFileAction.setShortcut('Ctrl+N')
        newFileAction.triggered.connect(self.newFile)
        
        openFileAction = QAction("Otworz plik", self)
        openFileAction.setShortcut('Ctrl+O')
        openFileAction.triggered.connect(self.openFile)
        
        saveFileAction = QAction("Zapisz plik", self)
        saveFileAction.setShortcut('Ctrl+S')
        saveFileAction.triggered.connect(self.saveFile)
        
        undoAction = QAction("Cofnij", self)
        undoAction.setShortcut('Ctrl+Z')
        undoAction.triggered.connect(self.undo)

        redoAction = QAction("Powtorz", self)
        redoAction.setShortcut('Ctrl+Y')
        redoAction.triggered.connect(self.redo)
        
        cutAction = QAction("Wytnij", self)
        cutAction.setShortcut('Ctrl+X')
        cutAction.triggered.connect(self.cut)
        
        copyAction = QAction("Kopiuj", self)
        copyAction.setShortcut('Ctrl+C')
        copyAction.triggered.connect(self.copy)
        
        pasteAction = QAction("Wklej", self)
        pasteAction.setShortcut('Ctrl+V')
        pasteAction.triggered.connect(self.paste)
        
        menubar = self.menuBar()
        fileMenu = menubar.addMenu('&Plik')
        fileMenu.addAction(newFileAction)
        fileMenu.addAction(openFileAction)
        fileMenu.addAction(saveFileAction)
        fileMenu.addAction(exitAction)
        editMenu = menubar.addMenu('&Edycja')
        editMenu.addAction(undoAction)
        editMenu.addAction(redoAction)
        editMenu.addAction(cutAction)
        editMenu.addAction(copyAction)
        editMenu.addAction(pasteAction)
       
        #Toolbar
        exitAct = QAction(QIcon('download.png'), 'Exit', self)
        exitAct.setShortcut('Ctrl+Q')
        exitAct.triggered.connect(qApp.quit)
        exitAct.setStatusTip('Quit')
       
       
        toolbar = self.addToolBar('Pierwszy')
        toolbar.addAction(QIcon('newwfile.png'), 'Nowy plik', self.newFile)
        toolbar.addAction(QIcon('openfile.png'), 'Otworz plik', self.openFile)
        toolbar.addAction(QIcon('search.png'), 'Szukaj pliku')
        toolbar.addAction(QIcon('save.png'), 'Zapisz pliku', self.saveFile)
        toolbar2 = self.addToolBar('Drugi')
        toolbar2.addAction(QIcon('back.png'), 'Cofnij', self.undo)
        toolbar2.addAction(QIcon('redo.png'), 'Powtorz', self.redo)
        toolbar3 = self.addToolBar('Trzeci')
        toolbar3.addAction(QIcon('cut.png'), 'Wytnij', self.cut)
        toolbar3.addAction(QIcon('copy.png'), 'Kopiuj', self.copy)
        toolbar3.addAction(QIcon('paste.png'), 'Wklej', self.paste)
       
        self.textEdit2.setText("przykladowy text")
       
        Widget.setLayout(grid)
        self.setGeometry(500, 500, 500, 300)
        self.setWindowTitle('Main window')
        self.statusBar().showMessage('Message in statusbar.')
        self.show()

if __name__ == '__main__':
    import sys
    app = QApplication(sys.argv)
    app.aboutToQuit.connect(app.deleteLater)
    okno = Kalkulator()
    sys.exit(app.exec())
```
