```python
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
#from PyQt5.QtWidgets import QLabel
# Filename: g_layout.py


class Kalkulator(QMainWindow):
    def __init__(self):
        super(Kalkulator, self).__init__()
        self.initUI()
    def initUI(self):
        Widget = QWidget()
        self.setCentralWidget(Widget)
       
        grid = QGridLayout()
       
        malyWidget = QWidget()
        malygrid = QGridLayout()
        button1 = QPushButton()
        button1.setIcon(QIcon("download.png"))
        button2 = QPushButton()
        button2.setIcon(QIcon("ikonka.png"))
        button3 = QPushButton()
        button3.setIcon(QIcon("ikonka2.png"))
        button1.setSizePolicy(QSizePolicy.Fixed,QSizePolicy.Fixed)
        button2.setSizePolicy(QSizePolicy.Fixed,QSizePolicy.Fixed)
        button3.setSizePolicy(QSizePolicy.Fixed,QSizePolicy.Fixed)
        button1.setFixedSize(30,30)
        button2.setFixedSize(30,30)
        button3.setFixedSize(30,30)
        malygrid.addWidget(button3,0,0)
        malygrid.addWidget(button2,0,1)
        malygrid.addWidget(button1,0,2)
        malyWidget.setSizePolicy(QSizePolicy.Fixed,QSizePolicy.Fixed)
        malyWidget.setLayout(malygrid)
        grid.addWidget(malyWidget,0,0)
       
        #grid na duze okna
        oknawidget = QWidget()
        oknagrid = QGridLayout()
        model = QFileSystemModel()
        model.setRootPath('This PC')
        textEdit = QTreeView(Widget)
        textEdit.setModel(model)
        textEdit2 = QTreeView(Widget)
        textEdit2.setModel(model)
        oknagrid.addWidget(textEdit,0,0,3,1)
        oknagrid.addWidget(textEdit2,0,2,3,1)
        oknawidget.setLayout(oknagrid)
        grid.addWidget(oknawidget,1,0)
       
       
        #grid na sciezke
        sciezkaWidget = QWidget()
        sciezkagrid = QGridLayout()
        title = QLabel('c:/>')
        titleEdit = QLineEdit()
        #titleEdit.setSizePolicy(QSizePolicy.Expanding,QSizePolicy.Fixed)
        sciezkagrid.addWidget(title,0,0)
        sciezkagrid.addWidget(titleEdit,0,1)
        sciezkaWidget.setLayout(sciezkagrid)
        sciezkaWidget.setSizePolicy(QSizePolicy.Expanding,QSizePolicy.Fixed)
        grid.addWidget(sciezkaWidget,2,0)
       
        dolnyWidget = QWidget()
        dolnygrid = QGridLayout()
        names = ['F3 View','F4 Edit','F5 Copy','F6 Move','F7 New Folder','F8 Delete','Alt+F4 Exit']
        j = 0
        for i in names:
            button = QPushButton(i)
            button.setSizePolicy(QSizePolicy.Expanding,QSizePolicy.Fixed)
            dolnygrid.addWidget(button, 0, j)
            j = j + 1
        dolnyWidget.setSizePolicy(QSizePolicy.Expanding,QSizePolicy.Fixed)
        dolnyWidget.setLayout(dolnygrid)
        grid.addWidget(dolnyWidget,3,0)
       
        exitAction = QAction('Exit', self)
        exitAction.setShortcut('Ctrl+Q')
        exitAction.setStatusTip('Exit application')
        exitAction.triggered.connect(self.close)
        self.statusBar()
        menubar = self.menuBar()
        fileMenu = menubar.addMenu('&File')
        fileMenu = menubar.addMenu('&Mark')
        fileMenu = menubar.addMenu('&Commands')
        fileMenu = menubar.addMenu('&Net')
        fileMenu = menubar.addMenu('&Show')
        fileMenu = menubar.addMenu('&Configuration')
        fileMenu = menubar.addMenu('&Start')
        fileMenu = menubar.addMenu('&Help')
        fileMenu.addAction(exitAction)
        Widget.setLayout(grid)
        self.setGeometry(300, 300, 250, 250)
        self.setWindowTitle('Total commander in python')
        self.setWindowIcon(QtGui.QIcon('icon.png'))
        self.show()

if __name__ == '__main__':
    import sys
    app = QApplication(sys.argv)
    app.aboutToQuit.connect(app.deleteLater)
    okno = Kalkulator()
    sys.exit(app.exec())
```
